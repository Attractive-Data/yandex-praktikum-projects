```python
# Создадим талблицу для времянного ряда с точки зрения распределения (начиная с 2009 года и заканчивая 2012-06-01,
# тк далее очень мало наблюдений)

time_series = """

WITH f AS
(SELECT CAST(DATE_TRUNC('MONTH', COLLISION_DATE) as DATE) AS date,
       COUNT(CASE_ID)
FROM collisions
WHERE COLLISION_DATE BETWEEN '2009-01-01' AND '2012-06-01'
GROUP BY CAST(DATE_TRUNC('MONTH', COLLISION_DATE) as DATE))

SELECT EXTRACT(MONTH FROM date) AS accidents_date,
       AVG(count) AS accidents_mean
FROM f
GROUP BY EXTRACT(MONTH FROM date);
"""

time_series = pd.read_sql_query(time_series, con = engine)

# Создадим словарь для месяцев
month_list = {1 : 'Январь',
              2 : 'Февраль',
              3 : 'Март',
              4 : 'Апрель',
              5 : 'Май',
              6 : 'Июнь',
              7 : 'Июль',
              8 : 'Август',
              9 : 'Сентябрь',
              10 : 'Октябрь',
              11 : 'Ноябрь',
              12 : 'Декабрь'} 

# Отсортируем и сбросим индексы
time_series = time_series.sort_values(by='accidents_date').reset_index(drop=True)

# Поменяем int на месяцы
time_series['accidents_date'] = time_series['accidents_date'].replace(month_list)

# Поставим месяцы на место индексов
time_series = time_series.set_index('accidents_date')
time_series
```

```python
# Изобразим наш Временной ряд

fig, ax = plt.subplots()
fig.set_figwidth(15)
fig.set_figheight(7)
ax.plot(time_series.index, time_series['accidents_mean'])
ax.set_title('Временной ряд ДТП', fontsize=20)
ax.set_xlabel('Месяц проишествия')
ax.set_ylabel('Среднее количество проишествий')
plt.show()
```

Сразу добавим в таблицу только те факторы, которые могут влиять на вероятность дтп:

```python
query = '''
with 
  p as(
        select distinct case_id,
               at_fault,
               party_type, 
               party_sobriety,
               party_drug_physical,
               cellphone_in_use
        from parties),
  c as (
        select distinct case_id,
               weather_1,
               road_surface,
               lighting,
               road_condition_1,
               collision_date,
               collision_damage,
               intersection,
               control_device
        from collisions),
  v as (
        select distinct case_id,
               vehicle_age
        from vehicles),
  c_i as (
          select distinct case_id
          from case_ids)
select p.party_type,
       p.at_fault,
       p.party_sobriety,
       p.party_drug_physical,
       c.road_surface,
       c.lighting,
       c.road_condition_1,
       v.vehicle_age,
       c.intersection,
       p.cellphone_in_use,
       c.control_device,
       c.weather_1
from c_i
join p on c_i.case_id = p.case_id
join c on c_i.case_id = c.case_id
join v on c_i.case_id = v.case_id
where party_type = 'car'
  and collision_date between '2012-01-01' and '2013-01-01'
  and collision_damage != 'scratch'
'''
df = pd.read_sql_query(query, con=engine)
```

## Проведите анализ важности факторов ДТП

Precision можно интерпретировать как долю объектов, названных классификатором положительными и при этом действительно являющимися положительными, а recall показывает, какую долю объектов положительного класса из всех объектов положительного класса нашел алгоритм. Recall демонстрирует способность алгоритма обнаруживать данный класс вообще, а precision — способность отличать этот класс от других классов. Precision и recall не зависят, в отличие от accuracy, от соотношения классов и потому применимы в условиях несбалансированных выборок. В идеале, стоит задача найти оптимальный баланс между этими двумя метриками, поэтому в данной работе я буду использовать метрику F1

### Decision Tree Classifier

```python
dt_params = {
    'max_depth': [8],
    'min_samples_split': [5], 
    'min_samples_leaf': [25],
    'class_weight':['balanced']
}

dt_gs = GridSearchCV(DecisionTreeClassifier(random_state=5),
                         dt_params, n_jobs=-1,cv = 5,scoring = 'f1')
dt_f1 = dt_gs.fit(features_train, target_train)
```

```python
print('Оптимальные параметры: ',dt_f1.best_params_) 
```

```python
print(f'Лучший roc_auc на обучающей выборке: {dt_f1.score(features_train, target_train)}')
print(f'Лучший roc_auc на тестовой выборке: {dt_f1.score(features_test, target_test)}')
```

### Random Forest Classifier

```python
best_depth = 0
best_accuracy = 0

for depth in range(1,16):
    for est in range(5,51,5):
        for sample in range(2,5):
            model = RandomForestClassifier(max_depth=depth, n_estimators=est, 
                                           min_samples_leaf=sample, random_state=12345)
            model.fit(features_train, target_train)
            accuracy = model.score(features_test, target_test)
            if accuracy > best_accuracy:
                best_accuracy = accuracy
                best_depth = depth
                best_est = est
                best_sample = sample
print("Лучшая глубина дерева:", best_depth)
print("Лучшее количество деревьев:", best_est)
print("Лучший минимум объектов в узле:", best_sample)
print("Accuracy:", best_accuracy)
```

```python
rf_params = {
    'criterion':['entropy'],
    'n_estimators': [45],
    'max_depth': [6],
    'min_samples_leaf': [3],
    'class_weight':['balanced']
}

rf_gs = GridSearchCV(RandomForestClassifier(random_state=12345),
                     rf_params, cv = 10, scoring = 'f1',
                     verbose=2)
rf_f1 = rf_gs.fit(features_train, target_train)
```
